# **Decision Tree**

### **1. Introduction to Decision Tree**

- **Decision Trees** are versatile, non-parametric supervised learning algorithms used for both classification and regression tasks. They work by recursively splitting the dataset into subsets based on the most significant feature at each node, forming a tree-like structure.
- ![Decision_Tree](./Images/1.png)
- **Internal nodes** represent feature comparisons (conditions).
- **Branches** represent the outcomes of these conditions.
- **Leaf nodes** represent the final prediction (class or value).
- ![Decision_Tree](<./Images/3%20(1).png>)

#### **Key Concepts**

- **Root Node** The starting point of the tree that contains all the data.
- **Decision Node** A node that splits the data based on certain features.
- **Leaf Node** The final output of the tree that makes a classification or prediction.

#### **Advantages**

- Simple to understand and interpret.
- Can handle both numerical and categorical data.
- Requires little data preprocessing.
- Can model non-linear relationships.

#### **Disadvantages**

- Prone to overfitting, especially with deep trees.
- Sensitive to small variations in the data.
- May create biased trees if some classes dominate the data.

### **3. Implementation of Decision Tree using Python**

#### **Import libraries and dataset**


```python
import numpy as np
import pandas as pd
df = pd.read_csv('Iris.csv')
df.head()
```

#### **Seperating the training set and test set**

```python
from sklearn.model_selection import train_test_split
x = df.drop('Species', axis = 1)
# x = df.drop('Id', axis=1)

y = df['Species']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30)
x_train2, x_validation, y_train2, y_validation = train_test_split(x_train,y_train,test_size=0.30)
```

#### **Importing Decision Tree Classifier via sklearn**

```python
from sklearn.tree import DecisionTreeClassifier
```

#### **Applying the Decision tree algorithm using entropy**

```python
dtree = DecisionTreeClassifier(criterion='entropy', max_depth=3)
dtree.fit(x_train2,y_train2)
predictions = dtree.predict(x_validation)
# predictions2 = dtree.predict(x_test)
```

#### **Checking the accuracy**

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_validation, predictions)
# predictions
```

#### **Applying the decision tree algorithm using gini**

```python
dtree2 = DecisionTreeClassifier(criterion='gini', max_depth=3)
dtree2.fit(x_train2,y_train2)
predictions = dtree2.predict(x_validation)
# predictions2 = dtree2.predict(x_test)
```

#### **Checking the accuracy**

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_validation, predictions)
```

#### **Plotting**

```python
import matplotlib.pyplot as plt 
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(18,8))
a = plot_tree(dtree, feature_names = df.columns, fontsize = 12, filled = True, class_names = ['Setosa', 'Versicolor', 'Virginica'])
```

```python
fig = plt.figure(figsize=(18,8))
a = plot_tree(dtree2, feature_names = df.columns, fontsize = 12, filled = True, class_names = ['Setosa', 'Versicolor', 'Virginica'])
```
