## **1. Introduction to Clustering and K-Means Algorithm**

- ### **What is Clustering?**

  - **Clustering** is a type of **unsupervised learning** where the aim is to group a set of objects in such a way that objects in the same group (cluster) are more similar to each other than to those in other groups.
  - Useful for pattern detection in data when labels are unavailable (e.g., customer segmentation or image region detection).
  - Though sometimes it may give gibberish.

  - ![Clustering Example](<./Images/Clustering%20(6).png>)

#### **K-Means Clustering**

- A commonly used algorithm that partitions the data into **K clusters** by iteratively updating centroids. -![KMeans Example](<./Images/Clustering%20(7).png>) - ![KMeans Example](<./Images/Clustering%20(8).png>)

#### **Steps of K-Means Algorithm**

1. Choose the number of clusters, K.
2. Randomly initialize K centroids.
3. Repeat until convergence:
   - Assign each data point to the nearest centroid.
   - Recalculate centroids based on the new clusters.
     - ![KMeans Example](<./Images/Clustering%20(9).png>)
       - ![KMeans Example](<./Images/Clustering%20(10).png>)

##### **Advantages**

- Simple and scalable.
- Efficient for large datasets.

##### **Disadvantages**

- May lead to suboptimal clusters if the initial centroids are poorly chosen.
- Sensitive to outliers and noise.

## **2. Step-by-step Example of K-Means Algorithm**

### **Numerical on K-Means Clustering**

- 1. **Initialize** centroids randomly.
- 2. Calculate the **distance** of each point to the centroid.
- 3. **Assign** each point to the nearest centroid.
- 4. **Recompute** the centroids based on the new clusters.
  - #### **The Algorithm**
    - In the beginning we determine number of cluster K and we assume the centroid or center of these clusters. We can take any random objects as the initial centroids or the first K objects in sequence can also serve as the initial centroids.
      - Then, the K Means algorithm will do the three steps below until convergence.
      - Iterate until stable(=no object move group):
        1.  Determine the centroid coordinate
        2.  Determine the distance of each object to the centroids
        3.  Group the object based on minimum distance.

## **3. Implementation of K-Means in Python**

#### **K-Means Clustering**

#### **Lets start by visualizing some data points**


```python
import matplotlib.pyplot as plt 
# Data points
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

# Scatter plot of data
plt.scatter(x,y)
plt.show()
```

#### **What should be the ideal value of k? - Elbow Method**
**Elbow method to find the optimal K**

```python
from sklearn.cluster import KMeans

data = list(zip(x,y))
intertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    intertias.append(kmeans.inertia_)

plt.plot(range(1,11), intertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Intertia/Centroids')
plt.show()
```

#### **Construction of the model**
**K-Means clustering**

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x,y, c=kmeans.labels_)
plt.show()
```

## **4. Introduction to Hierarchical Clustering**

#### **Hierarchical Clustering**

- Produces nested clusters that can be visualized as a **dendrogram**.
- Produces a set of nested clusters organized as a hierarchical tree
- Can be visualized as a dendrogram

  - Dendrogram is a tree-line diagram that records the sequences of merges or splits
  - ![Hierarchical Clustering](<./Images/Clustering%20(20).png>)
  - ##### **Strengths of Hierarchical Clustering**
    - No assumptions on the number of clusters
      - (Any desired number of clusters can be obtained by 'cutting' the dendrogram at the proper level.)
      - Does not require pre-specification of the number of clusters.
      - Can produce interpretable taxonomies.
    - Hierarchical clusterings may correspond to meaningful taxonomies.
      - (Eg: Phylogeny reconstruction, product catalogs, etc.)
  - ![Hierarchical Clustering](<./Images/Clustering%20(21).png>)

  - ##### **Complexity of Hierarchical Clustering**

    - Distance matrix is used for deciding which clusters to merge/split
    - At least quadratic in the number of data points
    - Not usable for large datasets
      - **Weaknesses**
        - Computationally expensive for large datasets.
        - Requires a **distance matrix**, making it quadratic in complexity.

  - ##### **Agglomerative Clustering Algorithm**

    - Most popular hierarchical clustering technique
    - Basic Algorithm:
      - Compute the distance matrix between the input data points
      - Let each data point be a cluster
      - Repeat
        - Merge the two closest cluster
        - Update the distance matrix
      - Until only a single cluster remains
    - Key operation is the computation of the distance between two clusters - Different definitions of the distance between clusters lead to differnt algorithms.

## **5. Implementation of Hierarchical Clustering in Python**
#### **Lets start by visualizing some data points**

```python
import matplotlib.pyplot as plt 

# Data points
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

# Scatter plot
plt.scatter(x,y)
plt.show()
```

#### **We compute the ward linkage using euclidean distance, and visualize it using a dendogram.**

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage


# Ward linkage and dendrogram
data = list(zip(x,y))
linkage_data = linkage(data,method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

### **We visualize the clusters on a scatter plot**
- **hierarchial_cluster = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage='ward')**
- **The error you're encountering is due to the fact that in recent versions of scikit-learn, the affinity parameter has been replaced by metric when using the AgglomerativeClustering class with the linkage='ward' option.**

```python
from sklearn.cluster import AgglomerativeClustering

# Hierarchical clustering
hierarchial_cluster = AgglomerativeClustering(n_clusters = 2, metric='euclidean', linkage='ward')
labels= hierarchial_cluster.fit_predict(data)

plt.scatter(x,y,c=labels)
plt.show()
```
