# **K-Nearest Neighbours(KNN)**

### **1. Introduction to K-Nearest Neighbours (KNN)**

- **KNN** is a simple and powerful supervised machine learning algorithm used primarily for classification. It classifies data points based on the majority vote of their nearest neighbors.
- ![KNN](<./Images/KNN%20(1).png>)

- **How KNN Works**

  - The distance between data points is calculated (commonly using Euclidean distance).
  - The query point is assigned to the class with the most representatives among its k nearest neighbors.

- **Choosing the value of k**
  - The value of k plays a crucial role in the model's performance. A small k can lead to overfitting, while a large k can lead to underfitting. The optimal k is usually found by testing different values and plotting accuracy or error rates against them.

### **2. Implementation of KNN using Python**

#### **Import the dataset**


```python
import numpy as np  # Importing for numerical operations on arrays.
import pandas as pd  # Importing for data manipulation and analysis.
import matplotlib.pyplot as plt  # Importing for creating visualizations/plots.
from sklearn.datasets import make_blobs  # to generate synthetic datasets for clustering.
from sklearn.neighbors import KNeighborsClassifier  # Importing KNeighborsClassifier for the k-nearest neighbors algorithm.
from sklearn.model_selection import train_test_split  # Importing train_test_split for splitting data into training and testing sets.
from sklearn.metrics import accuracy_score
```

#### **Creating the dataset**

```python
x,y = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1.5, random_state=4)
```

#### **Visualizing the dataset**

```python
# plt.style.use('seaborn')
plt.style.use('ggplot')  # Using a different valid style
plt.figure(figsize=(10,10))
plt.scatter(x[:,0],x[:,1], c=y ,marker='*',s=100,edgecolors='black')
plt.title('Generated Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

#### **Splitting the dataset into training and testing dataset**

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

#### **Testing various values of k**

```python
k_values = range(1, 21)
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred) * 100
    accuracies.append(accuracy)
    print(f"Accuracy with k={k}: {accuracy:.2f}%")
```

#### **Plotting the accuracy vs k**

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title('K-Nearest Neighbors Accuracy vs. k')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy (%)')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

**Performance** The K-Nearest Neighbors (KNN) model shows varying accuracy with different values of **k**. The accuracy improves as **k** increases, reaching a peak at **k = 19** with 96.67% accuracy.
