# **Naive Bayes**
## **14. Introduction to Naive Bayes Classifier**
- **Naive Bayes Classifier** is a simple, yet powerful supervised machine learning algorithm used for classification tasks. It is based on **Bayes' Theorem**, which provides a way to calculate the probability of a hypothesis based on prior knowledge.
  
  - **Bayes' Theorem** formula
    $$
    P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
    $$
  - Here, **P(A|B)** is the probability of event A occurring given that B is true. **P(B|A)** is the probability of event B occurring given that A is true. **P(A)** and **P(B)** are the independent probabilities of A and B, respectively.

- **Why is it called Naive?** The classifier is termed "naive" because it assumes that the features in the dataset are independent of each other, which is often not the case in real-world scenarios. Despite this strong assumption, it often performs well in practice.

## **Working of Naive Bayes Classifier**
- ![Naive Bayes2](./Images/Naive%20Bayes%20(2).png)
1. **Convert the dataset into frequency tables** This step involves calculating the frequency of each class label for every feature in the dataset.
- ![Naive Bayes3](./Images/Naive%20Bayes%20(3).png)
2. **Generate likelihood tables** Compute the likelihood of each feature value given a class label.
- ![Naive Bayes4](./Images/Naive%20Bayes%20(4).png)
3. **Apply Bayes' Theorem** Use the likelihood tables to calculate the posterior probability for each class label given the input features, and predict the class with the highest posterior probability.

### **2. Implementation of Naive Bayes Classifier using sklearn**
#### **Import the dataset**


```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

#### **Load dataset**

```python
iris = load_iris()
print(iris.target_names)

x, y = load_iris(return_X_y=True)
```

#### **Split the dataset**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=0)
```

#### **Fit the model**

```python
gnb = GaussianNB()
a = gnb.fit(x_train, y_train)
a
```

#### **Make predictions using the model**

```python
y_pred = gnb.predict(x_test)
```

#### **Print the accuracy or Naive Bayes Score**

```python
print("Naive Bayes Score:", gnb.score(x_test, y_test))
```

#### **Displaying the confusion matrix**

```python
cf = confusion_matrix(y_test, y_pred)
print(cf)
sns.heatmap(cf, annot=True)
```

**Performance** A model with a Naive Bayes score of 94.7% indicates strong predictive accuracy.
